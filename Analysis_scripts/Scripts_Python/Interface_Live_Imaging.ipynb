## Interface dynamics (first two moments)
We now determine how the interface height and width evolve over time. <br>

<u>Fisher waves</u><br>
Recall that the Fisher–KPP equation is <br>
    $\hspace{1cm} \partial_t u = D \partial_{xx} u + f(u), \hspace{1cm} f(u) = k u (1-u) $ <br>
For a Fisher wave, the interface grows linearly with time, $\langle h(t) \rangle = v t$. The deterministic model predicts $v = 2 \sqrt{D f'(0)} = 2 \sqrt{D k}$. The stochastic model based on $A \leftrightarrow A + A$ predicts $v = Dk$ in $d=1$ (Moro, PRL 2001). <br>

For the width, different scaling relations for before the saturating regime have been obtained (recall $w(t) \sim t^{\beta}$). Riordan et al. (PRL, 1995) do numerically study the stochastic process on a 2D grid and obtain diffusive scaling, i.e. $\beta=1/2$. Moro (PRL, 2001) shows that with a different interface definition, $\beta = 0.27 \pm 0.01$. 

<u>Other interface growth models</u> <br>
Tumor growth: linear expansion (Bru et al., PRL 1998 & BPJ 2003), also confirmed in CA model simulations by Block et al., 2007.
Monolayer Vero cell colony (Huergo et al., 2010): linear expansion of colony height. $\beta = 0.33$ for large colonies, but $\beta$ depends on system size $l$. 


### Method 1: intuitive method
Define the average interface height as $\langle h(t) \rangle = \frac{1}{L} \int h(y,t)\ dy$. <br>
Define the second moment $\langle h(t)^2 \rangle = \frac{1}{L} \int h(y,t)^2\ dy$. <br>
The width is then defined as $w(L, t) = \sqrt{\langle h(t)^2 \rangle - \langle h(t) \rangle^2}$.

#### (1) Mean interface height dynamics $\langle h(t) \rangle $

```python
heights_all_1 = np.nanmean(interfaces_all, axis=1)/pix_per_micron;
#heights_all_2 = np.nanmean(interfaces_all_2, axis=1)/pix_per_micron;
t_all = np.arange(t_max)/6;

h = plt.figure(dpi = 100);
plt.plot(t_all, heights_all_1); #, label='unblurred')
#plt.plot(t_all, heights_all_2, label='blurred')
plt.xlabel('Time (hours)')
plt.ylabel('Mean interface height <h(t)> ($\mu m$)')

# Fit linear function h = a*t + b, plot result
slope, intercept, r_value, p_value, std_err = stats.linregress(t_all, heights_all_1) 
plt.plot(t_all, slope*t_all + intercept, '--', label='$<h> \propto t$' + ', v = ' + str(round(slope, 2)) + '$\mu m/hr$')

# # Fit sqrt function h = b sqrt(a t)
def func(t, a, b):
    return a*np.sqrt(t) + b;
popt, pcov = optimize.curve_fit(func, t_all, heights_all_1)

plt.plot(t_all, popt[0]*np.sqrt(t_all) + popt[1], '--', label='$<h> \propto t^{1/2}$')

plt.legend();
#plt.xscale('log')
#plt.yscale('log')

# Save image
# fname_out = folder_figures + im_label + "interface_avg_height_dynamics_w_fits";
# label = '_method_1';
# plt.savefig(fname_out + label + ".png")
# plt.savefig(fname_out + label + ".pdf")

plt.show()
```

### Method 2: Probability density method (Podewitz et al., 2016)
Take $\langle h(t) \rangle = \int_{0}^{L_x} \varphi_X(x, t)\ dx$, where $\varphi_X(x, t)$ is the density of ON cells projected onto the x-axis. See Podewitz et al. (2016) and Riordan et al. (PRL, 1995). 

#### (1) Mean interface height dynamics

```python
# Do for a single image
phi_X = np.mean(data_image, axis=0); # data_image from before

# Calculate interface position
h = np.sum(phi_X)/pix_per_micron;
print(str(h))

# Calculate width
h2 = 2*np.sum( np.arange(len(phi_X))*phi_X )/(pix_per_micron**2); # height in pixels
width = np.sqrt(h2 - h**2);
print(width)

# Plot phi(X)
plt.figure();
plt.plot(range(len(phi_X)), np.flip(phi_X));
```

### Validity of scaling analysis

For the scaling arguments to work, the system needs to satisfy either of two conditions below.

#### (1) Family-Vicsek scaling
$\hspace{1cm} w(L,t) \sim L^{\alpha} f\left(\frac{t}{L^z}\right)$. <br>
Would we also verify this condition by looking at how the <i>local</i> width $w(l,t)$ scales?


#### (2) Fractal dimension
Alternatively, Bru et al. (2003) say that it is sufficient to check that the interface is fractal, by calculating the fractal dimensions and showing that it's a non-integer number.

### Method 1: calculate exponents $\alpha$, $\beta$ and $z$ from interface width
First, we can define a length-scale-dependent interface width <br>
$\hspace{1cm} w(l, t) =  \left( \frac{1}{l}  \sum_{h_i \in l} \left[ h_i(t) - \langle h_i \rangle_l \right]  \right)_L^{1/2}$ <br>

This quantity is proposed to have scaling relations <br>
$\hspace{1cm} w(l, t) \propto \begin{cases} t^{\beta} & \text{if } t \ll t_s \\ l^{\alpha_{loc}} & \text{if } t \gg t_s \end{cases} $ <br>

Note that for $l < L$, the scaling relation gives the <i>local</i> roughness exponent $\alpha_{loc}$, which may differ from the global roughtness exponent $\alpha$.

Hence, we obtain 
<ol>
    <li>$\alpha_{loc}$ from the long-time scaling with $l$.</li>
    <li>$\beta$ from the short-time scaling with $t$. However, this only works if we have data from the initial phase before the width saturates. </li>
    <li>$z$ from the relation $z = \alpha/\beta$ (if we have $\beta$), or from the Family-Vicsek scaling, by plotting $w(l,t)/l^{\alpha}$ against $\frac{t}{l^z}$. For the correct values of $\alpha$ and $z$, the curves for different $l$ should collapse onto one curve. </li>
</ol>

```python
# calculate local width as function of l
def calc_local_width(data, l):
    var_all = [np.var(data[i:i+l]) for i in range(len(data)-l)];
    return np.mean(var_all)**0.5

#l_list = np.arange(1, len(interface_0)+1); 
l_list = np.unique(np.array( np.round(10**np.arange(0.1, 3.5, 0.1)), dtype=int ));
t_list = np.arange(t_max);
local_widths_all = [];

if np.max(l_list) > min(interface_lengths):
    print("WARNING! Value of l larger than interface length detected!");
    
for t in t_list:
    if t%10==0:
        print('t={}'.format(t));
    # load interface
    interface = interfaces_all[t];
    
    # process data: remove nan
    interface = interface[~np.isnan(interface)]; # TD: optimize speed
    #print(len(interface))
    
    # calculate local widths
    local_widths_temp = [calc_local_width(interface, l) for l in l_list];
    local_widths_all.append(local_widths_temp);

# convert result to np array
local_widths_all = np.array(local_widths_all);
print("Finished");
```

### Method 2: calculate exponents $\alpha$ and $z$ from structure factor
We can also the scaling exponents from the dynamic structure factor <br>
$\hspace{1cm} S(k, t) = \langle \tilde{h}(k, t) \tilde{h}(-k, t) \rangle = k^{-(2\alpha + 1)} \tilde{s}(k t^{1/z}),$ <br>
$\hspace{1cm} \tilde{s}(u = kt^{1/z}) = \begin{cases} u^{-(2\alpha+1)} &\text{if } u \ll 1 \\ \text{const.} &\text{if } u \gg 1 \end{cases} $ <br> 

Here $\tilde{h}(k, t)$ denotes the Fourier transform of the interface height function $h(y, t)$ and $\langle \ldots \rangle$ the average over many instances (i.e. images). <br>

1. Obtain $\alpha$.
Tor sufficiently large $k$, or at sufficiently large times $t$, $S(k, t)$ should scale as a power law and we can obtain the roughness coefficient $\alpha$ from this power law. <br>

2. Obtain $z$. 
Plot $S(k, t) k^{(2\alpha + 1)} = \tilde{s}(u)$ against $u = k t^{1/z}$. For the correct value of $z$, the curves from different $t$ should fall on top of each other (otherwise they split for small values of $u$). <br>

<b>Note:</b>  This approach is taken in these papers. However, it's not clear at all to me how they obtain an increasing function for small $u$, as the slope should be $-(2\alpha+1) < 0$...

We take the first coordinate at which the contiguous region is reached from the direction perpendicular to the interface. The result is that we can then define an interface height function $h(y,t)$.

```python
# Check single image
t = 0;
data_path = "/Users/dang/Documents/TablerLab/Data_Live_Imaging/190314 Live Imaging/Processed_TXT_data_interface/";
filename_in = "MaxProjected_Stitched_dataset(0-12hr)_YD_processed_interface_t" + str(t) + ".txt";
fullname_in = data_path + filename_in;

data_image = pd.read_csv(fullname_in, header = None, delimiter='\t');
data_image[data_image == 255] = 1; # convert to 0/1

# Store data shape
Lx = data_image.shape[1];
Ly = data_image.shape[0];

# Show loaded figure
plt.figure(dpi=100);
plt.imshow(data_image, cmap='gray', vmin=0, vmax=1)
plt.show()
```

```python
# If the "ON" cells are on the left of the image, reverse image (for analysis)
reverse = 1;

data_image_rev = np.empty(data_image.shape);
for i in np.arange(0, data_image.shape[0]):
    # print(i)
    data_image_rev[i] = data_image.iloc[i][::-1];
    
# Show reversed figure
#plt.figure(dpi=100);
#plt.imshow(data_image_rev, cmap='gray', vmin=0, vmax=1)
#plt.show()

# Define interface
if reverse:
    interface = np.array(data_image_rev.argmax(axis=1), dtype=float); # find (first) index of max. of data frame
else:
    interface = np.array(data_image.idxmax(axis=1), dtype=float); # find (first) index of max. of data frame
    interface = Lx - interface; # invert height if bone at right 

# Check that none of the rows has only 0 or 1:
if len(interface[interface == 0])==0:
    print("No empty/full rows")    
else:
    print("Empty/full rows detected!") 
    # post-processing: delete empty/full rows
    mask = np.array(interface == 0, dtype=bool);
    interface[mask] = np.nan; # remove bad rows
    
# reverse back for further use
interface = Lx - interface;
```

```python
y_vals = np.arange(interface.size); # normal y values

# Plot interface
plt.figure();
plt.plot(interface, y_vals)
plt.xlim(0, Lx)
plt.ylim(0, Ly)
plt.gca().invert_yaxis()

# Save image
fname_out = folder_figuressave_folder + label_0 + label_0 + "_interface";
ext = ".png";
#plt.savefig(fname_out + ext)
```
