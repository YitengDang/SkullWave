```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Import saved data
#folder = '/Users/dang/Documents/Projects/Tabler_skull/Scripts/Scripts_mathematica/Plots_model4b_FD_solver/simulation10/'
folder = '/Users/ydang3/Documents/Projects_Dresden/Tabler_SkullWave/SkullWave/Model_simulation_scripts/Model_results/_best_revision_2024_07'
file_id_str = '2024-07-19_1806_'
fnames = ['data_parameters.csv', 'data_rho_1.csv', 'data_phi_1.csv', 'data_V_1.csv']

# show parameters
df=pd.read_csv( os.path.join(folder, file_id_str+fnames[0]), header=None, index_col=0 )
df
```

```python
L=float(df.loc['L'].values[0]); Nx = int(df.loc['Nx'].values[0]);
tmax = float(df.loc['tmax'].values[0]); T = int(df.loc['T'].values[0]);
```

```python
datas = []
datas.append( pd.read_csv( os.path.join(folder, file_id_str+fnames[1]), header=None ).T ) # rho
datas.append( pd.read_csv( os.path.join(folder, file_id_str+fnames[2]), header=None ).T ) # phi
datas.append( pd.read_csv( os.path.join(folder, file_id_str+fnames[3]), header=None ).T ) # V
```

```python
# common plot variables
dpi = 300;
#extent = (-L, L, 0, tmax)
int_method = 'spline16'; # interpolation
titles = ['$\\rho(x,t)$', '$\\phi(x,t)$', '$V(x,t)$']
cmap_ranges = [(0.005, 0.01), (0, 1), (0, 12)]
```

```python
# Adjust plot limits
#xlims = np.array([0, 200])
xlims = np.array([30, 170])
#xlims = np.array([60, 140])
#xlims = np.array([75, 125])
xlims_um = xlims*(2*L/Nx)-L

#tlims = np.array([0, 100])
tlims = np.array([0, 75])
tlims_hr = tlims*(tmax/T)
tlims_hr

extent = (xlims_um[0], xlims_um[1], tlims_hr[0], tlims_hr[1])
```

```python
# Plotting parameters
plt.rcParams.update({'font.family':'Arial', 'font.size':14})
```

```python
# plot example
idx=1

h=plt.figure(dpi=dpi/3)
#plt.imshow(datas[idx], origin='lower', extent=extent, aspect='auto', interpolation=int_method)
plt.imshow(datas[idx].iloc[tlims[0]:tlims[1], xlims[0]:xlims[1] ], origin='lower', extent=extent, aspect='auto', 
           interpolation=int_method, vmin=cmap_ranges[idx][0], vmax=cmap_ranges[idx][1])
plt.xlabel('$x (\mu m)$')
plt.ylabel('$t$ (hours)')
plt.title(titles[idx])
plt.colorbar()

#fname_out = os.path.join(save_folder, save_fnames[idx]+'.pdf');
#plt.savefig(fname_out, format='pdf')
plt.show()
```

```python
# Take snapshot
ylims = [(0.005, 0.008), (-0.05, 1.05), (0, 14)]
ylabels = ['$\\rho\,\, (\mu m^{-2})$', '$\\phi$', '$V\,\, (\mu m/hr)$']

t_idx = int(T/2);
save_fnames=['plot_rho_t_idx_'+str(t_idx), 'plot_phi_t_idx_'+str(t_idx), 'plot_V_t_idx_'+str(t_idx)]

for idx in range(3):
    plot_data = datas[idx].iloc[ int(t_idx), ]

    plt.figure(dpi=dpi);
    plt.plot(np.linspace(-L, L, Nx+1), plot_data)
    plt.ylim(ylims[idx])
    plt.xlabel('$x\,\, (\mu m)$')
    plt.ylabel( ylabels[idx] )

    fname_out = os.path.join(save_folder, save_fnames[idx]+'.pdf');
    #print(fname_out)
    
    #plt.savefig(fname_out);
    plt.show()
```
